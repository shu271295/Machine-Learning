# Q Q Plot (Used to check type of distribution)

```python
# Q-Q plot is used to check distribution of random variable(X) by comparing it with other plot using "dist" parameter
```

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
```

```python
#How to generate random normal/gaussian sample => X ~ N(0,1)
std_normal_sample = np.random.normal(loc = 20, scale = 5 , size = 100)

stats.probplot(x = std_normal_sample, dist = "norm" , plot = plt)
```

# Observation

1) Points are lies close to straight line.

```python
uniform_sample = np.random.uniform(low = 1 , high = 100 , size = 1000)
stats.probplot(x= uniform_sample , dist = "norm" , plot = plt )
```
