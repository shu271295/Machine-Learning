```python
#PDF(Probability Density Function) and CDF(Cumulative Distribution Function)
#Using Numpy and matplotlib

#Setosa
counts,bin_edges = np.histogram(iris[iris.species == "setosa"].petal_length,bins=10,density=True)
pdf = counts/sum(counts)
cdf = np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf,"b")
plt.plot(bin_edges[1:],cdf,"b")

#Versicolor
counts,bin_edges = np.histogram(iris[iris.species == "versicolor"].petal_length,bins=10,density=True)
pdf = counts/sum(counts)
cdf = np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf,"r")
plt.plot(bin_edges[1:],cdf,"r")

#Verginica
counts,bin_edges = np.histogram(iris[iris.species == "virginica"].petal_length,bins=10,density=True)
pdf = counts/sum(counts)
cdf = np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf,"g")
plt.plot(bin_edges[1:],cdf,"g")

plt.xlabel("petal_length")
plt.ylabel("probability")
plt.show()
```

# Mean, Variance and Std Deviation

```python
#Means or Central Tendency
print("Means : ")
print("setosa = " + str(np.mean(iris[iris.species == "setosa"].petal_length)))
print("versicolor = " + str(np.mean(iris[iris.species == "versicolor"].petal_length)))
print("virginica = " + str(np.mean(iris[iris.species == "virginica"].petal_length)))

#Effect of outlier on mean
print("setosa with outlier = " + str(np.mean(np.append(iris[iris.species == "setosa"].petal_length, 50))))

#std-dev (standard deviation) or spread
print("\nStandard Deviation : ")
print("setosa = " + str(np.std(iris[iris.species == "setosa"].petal_length)))
print("versicolor = " + str(np.std(iris[iris.species == "versicolor"].petal_length)))
print("virginica = " + str(np.std(iris[iris.species == "virginica"].petal_length)))
```

# MEDIAN

```python
#Median <=> Mean
#Median (It will not easily corrupted with outlier)
print("Median : ")
print("setosa : " + str(np.median(iris[iris.species == "setosa"].petal_length)))
print("versicolor : " + str(np.median(iris[iris.species == "versicolor"].petal_length)))
print("virginica : " + str(np.median(iris[iris.species == "virginica"].petal_length)))

#Effect of outlier on median
# No Effect
print("setosa with outlier : " + str(np.median(np.append(iris[iris.species == "setosa"].petal_length, 50))))
```

# Quantiles

```python
#Quantiles are 0,25,50,75,100 percentiles
#Observations :
# 1) min petal_length of setosa flower = 1.0 & max petal_length = 1.9
# 2) 25 % of the flower have petal_length in range [1,1.4]
# 3) 50 % of the flower have petal_length in range [1,1.5] and likewise for 75 %
print(np.percentile(iris[iris["species"]=="setosa"].petal_length,np.arange(0,101,25)))

# we can pass either numpy array or list
print(np.percentile(iris[iris["species"]=="setosa"].petal_length,[0,25,50,75,100]))

# 90 th percentile
print(np.percentile(iris[iris["species"]=="setosa"].petal_length,90))
```

# MAD (Median Absolute Deviation)

```python
# MAD <=> std dev
# MAD = median ( |x - median(x) | )
# It will not easily corrupted with outlier
# Median function is not available in Numpy

from statsmodels import robust

print(robust.mad(iris[iris["species"] == "setosa"].petal_length))
```

# IQR (Inter Quartile Range)

```python
# It is a midspread or middle 50 % 
# IQR = (Q3 - Q1) or (75th percentile - 50th percentile)
```

# Box Plot with whiskers

```python
# It is another way of visualization for univariate analysis
# It considers Median, Percentile, Quantile, IQR
# Whiskers : No std method to draw whiskers.Whiskers can be drawn using min and max OR sometimes using 1.5*IQR

sns.boxplot(data = iris, x = "species" , y = "petal_length")

```
